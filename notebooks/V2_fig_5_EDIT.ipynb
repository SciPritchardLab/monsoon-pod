```python
def load(filename,filedir=FILEDIR):
    filepath = f'{filedir}/{filename}'
    ds = xr.open_dataset(filepath)
    return ds.load()
```

```python
def get_pdf(monthstats,bintype,precipitating=False,samplethresh=SAMPLETHRESH):
    if bintype=='1D':
        blbins = monthstats.bl.values
        Q0 = monthstats.Q0.values
        QE = monthstats.QE.values
        Q  = QE if precipitating else Q0
        norm = np.nansum(Q0)*np.diff(blbins)[0]
        pdf  = Q/norm
        return xr.DataArray(pdf,coords={'bl':blbins})
    if bintype=='2D':
        subsatbins = monthstats.subsat.values
        capebins   = monthstats.cape.values
        P0 = monthstats.P0.values
        PE = monthstats.PE.values
        P  = PE if precipitating else P0
        norm = np.nansum(P0)*np.diff(subsatbins)[0]*np.diff(capebins)[0]
        pdf  = P/norm
        return xr.DataArray(pdf,coords={'subsat':subsatbins,'cape':capebins})

def get_bin_mean_pr(monthstats,bintype,samplethresh=SAMPLETHRESH):
    if bintype=='1D':
        blbins = monthstats.bl.values
        Q0 = monthstats.Q0.values
        Q1 = monthstats.Q1.values
        Q0[Q0==0.0] = np.nan
        binmeanpr = Q1/Q0
        binmeanpr[Q0<samplethresh] = np.nan
        return xr.DataArray(binmeanpr,coords={'bl':blbins})
    if bintype=='2D':
        subsatbins = monthstats.subsat.values
        capebins   = monthstats.cape.values
        P0 = monthstats.P0.values
        P1 = monthstats.P1.values
        P0[P0==0.0] = np.nan
        binmeanpr = P1/P0
        binmeanpr[P0<samplethresh] = np.nan
        return xr.DataArray(binmeanpr,coords={'subsat':subsatbins,'cape':capebins})

def get_mode(pdf,bintype,idxs=False):
    if bintype=='1D':
        blbins    = pdf.bl.values
        maxblidx  = np.nanargmax(pdf.values)
        blmode    = blbins[maxblidx]
        return blmode
    if bintype=='2D':
        subsatbins = pdf.subsat.values
        capebins   = pdf.cape.values
        maxsubsatidx,maxcapeidx = np.unravel_index(np.nanargmax(pdf.values),pdf.values.shape)
        if idxs:
            return maxsubsatidx,maxcapeidx
        subsatmode,capemode = subsatbins[maxsubsatidx],capebins[maxcapeidx]
        return subsatmode,capemode
```

```python
def get_sensitivity_metrics(monthstats,samplethresh=SAMPLETHRESH):

    # Compute 2D precipitating PDF
    pdf = get_pdf(monthstats,bintype='2D',precipitating=True)
    
    # Get 2D bin mean precipitation "surface"
    binmeanpr = get_bin_mean_pr(monthstats,bintype='2D')
    
    # Find the SUBSAT index and CAPE index where the maximum PE value occurs
    maxsubsatidx,maxcapeidx = np.unravel_index(np.nanargmax(monthstats.PE.values),monthstats.PE.values.shape)

    # Loop through SUBSAT dimension of binmeanpr until a bin with non-NaN values is found
    while np.isnan(binmeanpr[:maxsubsatidx,:]).all():
        # Shift the SUBSAT index for maximum PE by 1 bin (drier)
        maxsubsatidx = maxsubsatidx+1
        # If we've gone beyond bounds of our data, raise an error
        if maxsubsatidx>monthstats.subsat.size-1:
            raise ValueError('Could not locate valid point for precipitation sensitivity computation.')

    # Divide binmeanpr into 3 distinct regions based on the location of (maxsubsatidx,maxcapeidx)
    # The precipitating PDF mode represents most common combination of CAPE/SUBSAT 
    # when P occurs. It serves as a good indicator of the atmospheric state that 
    # frequently leads to P in the given region/month. Using this mode makes it a
    # good reference point to see how P changes when you deviate from this typical
    # state.
    
    # Create 3 copies of the binmeanpr array filled with NaNs for our three regions
    binmeanpra = np.full_like(binmeanpr,np.nan)
    binmeanprb = np.full_like(binmeanpr,np.nan)
    binmeanprc = np.full_like(binmeanpr,np.nan)

    # Region C: conditions drier than typical (SUBSAT > max-1), with higher instability than typical (CAPE > max), 
    # How P changes as LFT dryness changes, while maintaining high instability
    # Role FLT dryness plays in driving convection when stability is not a limiting factor
    binmeanprc[maxsubsatidx-1:,maxcapeidx:] = binmeanpr[maxsubsatidx-1:,maxcapeidx:]

    # Region B: conditions moister than typical (SUBSAT < max), with lower instability than typical(CAPE < max+1)
    # How P changes as atmospheric stability changes, in relatively moist conditions
    # Role instability plays in driving convection when moisture is not a limiting factor
    binmeanprb[:maxsubsatidx,:maxcapeidx+1] = binmeanpr[:maxsubsatidx,:maxcapeidx+1]

    # Region A: conditions moister than typical (SUBSAT < max), with higher instability than typical (CAPE > max)
    # How P changes in highly unstable and moist environments
    # Understanding transition between CAPE-limited and moisture-limited convection regimes
    binmeanpra[:maxsubsatidx,maxcapeidx:]  = binmeanpr[:maxsubsatidx,maxcapeidx:]

    # Identify the non-NaN (finite) values in each region
    finitea = (np.where(np.isfinite(binmeanpra)))
    finiteb = (np.where(np.isfinite(binmeanprb)))
    finitec = (np.where(np.isfinite(binmeanprc)))

    # Extract CAPE/SUBSAT values from each region as necessary
    subsatavals = monthstats.subsat.values[finitea[0]] 
    subsatcvals = monthstats.subsat.values[finitec[0]]
    capeavals   = monthstats.cape.values[finitea[1]]
    capebvals   = monthstats.cape.values[finiteb[1]]

    # Get average CAPE/SUBSAT/precipitation values across regions
    subsata = subsatavals.mean()
    subsatc = subsatcvals.mean()
    capea = capeavals.mean()
    capeb = capebvals.mean()
    pra = np.nanmean(binmeanpra)
    prb = np.nanmean(binmeanprb) 
    prc = np.nanmean(binmeanprc) 

    # Get sensitivity metrics of interest
    capesensitivity   = abs(pra-prb)/abs(capea-capeb)
    subsatsensitivity = abs(pra-prc)/abs(subsata-subsatc)
    gammamag = np.sqrt(capesensitivity**2+subsatsensitivity**2)
    gamma = capesensitivity/subsatsensitivity
    theta = np.arctan(gamma)

    return gammamag,gamma,theta
    # return gammamag,gamma,theta,maxsubsatidx,maxcapeidx,subsata,capea
    return binmeanprc,binmeanprb,binmeanpra,maxsubsatidx,maxcapeidx
```
