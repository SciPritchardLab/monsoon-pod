```python
import warnings
import numpy as np
import xarray as xr
import proplot as pplt
pplt.rc.reso='hi'
warnings.filterwarnings('ignore')
```

```python
FILEDIR = '/global/cfs/cdirs/m4334/sferrett/monsoon-pod/data/processed'
SAVEDIR = '/global/cfs/cdirs/m4334/sferrett/monsoon-pod/figs'
CASES = {
    'JJ':[(6,'June','#D42028'),(7,'July','#F2C85E')],
    'JA':[(7,'July','#F2C85E'),(8,'August','#5BA7DA')]}
SAMPLETHRESH = 50
```

```python
def load(filename,filedir=FILEDIR):
    filepath = f'{filedir}/{filename}'
    ds = xr.open_dataset(filepath)
    return ds.load()
```

```python
hiresimergstats = load('ERA5_IMERG_binned_stats.nc')
loresimergstats = load('LOW_ERA5_IMERG_binned_stats.nc')
loresgpcpstats  = load('ERA5_GPCP_binned_stats.nc')
```

```python
def get_pdf(monthstats,bintype,precipitating=False,samplethresh=SAMPLETHRESH):
    if bintype=='1D':
        Q0 = monthstats.Q0.values
        QE = monthstats.QE.values
        Q  = QE if precipitating else Q0
        Q[Q0<samplethresh] = np.nan
        pdf = Q/np.nansum(Q)
        return xr.DataArray(pdf,coords={'bl':monthstats.bl.values})
    if bintype=='2D':
        P0 = monthstats.P0.values
        PE = monthstats.PE.values
        P  = PE if precipitating else P0
        # P[P0<samplethresh] = np.nan
        pdf = P/np.nansum(P)
        return xr.DataArray(pdf,coords={'subsat':monthstats.subsat.values,'cape':monthstats.cape.values})

def get_bin_mean_pr(stats,bintype='1D',samplethresh=SAMPLETHRESH):
    if bintype=='1D':
        coords = {'bl':stats.bl.values}
        Q0 = stats.Q0.values
        Q1 = stats.Q1.values
        Q0[Q0==0.0] = np.nan
        binmeanpr = Q1/Q0
        binmeanpr[Q0<samplethresh] = np.nan
    if bintype=='2D':
        coords = {'subsat':stats.subsat.values,'cape':stats.cape.values}
        P0 = stats.P0.values
        P1 = stats.P1.values
        P0[P0==0.0] = np.nan
        binmeanpr = P1/P0
        binmeanpr[P0<samplethresh] = np.nan
    return xr.DataArray(binmeanpr,coords=coords)

def get_sensitivity_metrics(monthstats,samplethresh=SAMPLETHRESH):
    binmeanpr = get_bin_mean_pr(monthstats,bintype='2D')
    maxsubsatidx,maxcapeidx = np.unravel_index(np.nanargmax(monthstats.PE.values),monthstats.PE.shape)
    while np.isnan(binmeanpr[:maxsubsatidx,:]).all():
        maxsubsatidx += 1
        if maxsubsatidx>=monthstats.subsat.size:
            raise ValueError('Could not locate valid point for precipitation sensitivity computation.')
    pra = np.copy(binmeanpr)
    prb = np.copy(binmeanpr)
    prc = np.copy(binmeanpr)
    pra[:] = np.nan
    prb[:] = np.nan
    prc[:] = np.nan
    pra[:maxsubsatidx,maxcapeidx:]   = binmeanpr[:maxsubsatidx,maxcapeidx:]
    prb[:maxsubsatidx,:maxcapeidx+1] = binmeanpr[:maxsubsatidx,:maxcapeidx+1]
    prc[maxsubsatidx-1:,maxcapeidx:] = binmeanpr[maxsubsatidx-1:,maxcapeidx:]
    maska = (np.where(np.isfinite(pra)))
    maskb = (np.where(np.isfinite(prb)))
    maskc = (np.where(np.isfinite(prc)))
    capea   = monthstats.cape.values[maska[1]]
    capeb   = monthstats.cape.values[maskb[1]]
    subsata = monthstats.subsat.values[maska[0]]
    subsatc = monthstats.subsat.values[maskc[0]]
    pramean = np.nanmean(pra)
    prbmean = np.nanmean(prb)
    prcmean = np.nanmean(prc)
    capeamean = capea.mean()
    capebmean = capeb.mean()
    subsatamean = subsata.mean()
    subsatcmean = subsatc.mean()
    dprcape   = abs(pramean-prbmean)
    dprsubsat = abs(pramean-prcmean)
    dcape     = abs(capeamean-capebmean)
    dsubsat   = abs(subsatamean-subsatcmean)
    gamma    = (dprcape/dcape)*(dsubsat/dprsubsat)
    gammamag = np.sqrt(((dprcape/dcape)**2)+((dprsubsat/dsubsat)**2))
    theta    = np.arctan(gamma)
    return theta,gamma,gammamag,subsatamean,capeamean
```

```python
def plot_dataset(stats,title):
    fig,axs = pplt.subplots(nrows=1,ncols=5)
    axs.format(suptitle=title,abcloc='l',abc='a)',xlabel='$SUBSAT_L$ (K)',xlim=(0,20),xticks=5,ylabel='$CAPE_L$ (K)',ylim=(-10,10),yticks=5)
    for i,region in enumerate(stats.region.values):
        axs[i].format(titleloc='l',title=f'{region}')
        casekey = 'JJ' if i<3 else 'JA'
        axnum   = 3 if casekey=='JA' else 0
        keylist = []
        for month,label,color in CASES[casekey]:
            monthstats = stats.sel(region=region,month=month)
            pdf = get_pdf(monthstats,bintype='2D',precipitating=False)
            maxsubsatidx,maxcapeidx = np.unravel_index(np.nanargmax(monthstats.P0.values),monthstats.P0.shape)
            subsatmode,capemode     = monthstats.subsat.values[maxsubsatidx],monthstats.cape.values[maxcapeidx]
            pdfmax     = np.nanmax(pdf)
            pdflevels  = np.arange(pdfmax*0.1,pdfmax,pdfmax*0.2)
            m = axs[i].contour(pdf.T,color=color,levels=pdflevels,label=label,alpha=0.7)
            keylist.append(m)
            axs[i].scatter(subsatmode,capemode,marker='*',s=100,edgecolor='k',linewidth=0.5,color=color,zorder=10)
        axs[axnum].legend(keylist,[CASES[casekey][j][1] for j in range(len(CASES[casekey]))],loc='ur',ncols=1)
    pplt.show()

plot_dataset(hiresimergstats,'ERA5/IMERG V06')
plot_dataset(loresimergstats,'LOW-ERA5/IMERG V06')
plot_dataset(loresgpcpstats,'ERA5/GPCP')
```
